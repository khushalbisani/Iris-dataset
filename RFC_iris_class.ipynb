#                      RandomForestClassifier on IRIS dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import operator

from sklearn.cross_validation import train_test_split
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier 
```

```python
iris=pd.read_csv(r"C:\Users\khush\Documents\Iris dataset\Iris.csv")
iris.drop('Id',axis=1,inplace=True)
iris.head(2)
```

```python
train, test = train_test_split(iris, test_size = 0.4)
print(train.shape)
print(test.shape)
```

```python
train_x=train[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
train_y=train.Species

test_x=test[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
test_y=test.Species
```

```python
model=RandomForestClassifier(n_estimators=10)
```

```python
model.fit(train_x,train_y)
```

```python
prediction=model.predict(test_x)
prediction
```

```python
acc=metrics.accuracy_score(prediction,test_y)*100
print('The accuracy of the RFC is',acc)
```
